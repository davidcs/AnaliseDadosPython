### Parte 2: Historico de Pedidos
Nessa parte queremos entender quais os melhores dias da nossa operação em faturamento. E a partir dai criar as hipoteses e planos de ações para que possamos repetir mais dias bons como esses.
Importamos o arquivo excel a qual continha todos os historicos de pedidos feitos no foodtruck. 

```python
historicoPedidos = pd.read_excel("Historico_Pedidos.xlsx")
display(historicoPedidos.head())
```

### Observação:
Apos listar todos os pedidos da operação. Vimos a necessidade de criar um filtro a qual possibilitasse ver quais eram os 10 melhores dias trabados no foodtruck.

Criamos inicialmente um filtro  a qual mostrava a soma de todos os pedidos do dia e o seu valor final. Porem o não nos mostrava os melhores dias. Os dias com maior faturamento. 

```python
# faturamento por dia
faturamento_por_dia = historicoPedidos[["Aberto", "Total Recebido"]].groupby(["Aberto"]).sum()
display(faturamento_por_dia.head())
```

### Filtro dos 10 melhores dia:
Abaixo ja o display a qual mostra o filtro que soma o faturamento dos dias trabalhados, agrupa eles e mostra em ordem crescente os 10 melhores dias trabalhados ate hoje no foodtruck.

```python
# faturamento dos 10 melhores dias
faturamento_por_dia = historicoPedidos[["Aberto", "Total Recebido"]].groupby(["Aberto"]).sum()
display(faturamento_por_dia.sort_values("Total Recebido", ascending=False).head(10))
```
